```python
class TempData:
    def __init__(self):
        self.data = [[[0]*800]*200]*2  #{0: [[0]*800]*200, 1: [[[0]*800]*200]}
        self.data = {0: [[0]*800]*200, 1: [[[0]*800]*200]}

    def append(self, idx, sig):
        self.data[idx].pop(0)

        self.data[idx].append(sig)
    def read(self, idx):
        return sum(self.data[idx], start=[])
mics = mic.JointMicrophoneReader([9, 10], 800)
x = TempData()
mics.start(x.append)
```
